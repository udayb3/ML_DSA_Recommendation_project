**Using Embedding**

Generating the word to vectors using the embedding , for preserving the better context of the sentences and can be used to recommend the other similar questions.

```python
from transformers import DistilBertModel, DistilBertTokenizer
import torch
import pandas as pd

# Check if GPU is available and load model to GPU if possible
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the smaller DistilBERT model and tokenizer
model_name = "distilbert-base-uncased"
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DistilBertModel.from_pretrained(model_name).to(device)

# Define a function to get DistilBERT embeddings in batches
def get_distilbert_embeddings_batch(texts, batch_size=16, max_length=64):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]

        # Encode the batch of texts
        inputs = tokenizer(batch_texts, return_tensors="pt", truncation=True, padding=True, max_length=max_length)
        input_ids = inputs['input_ids'].to(device)

        # Generate embeddings for the batch
        with torch.no_grad():
            outputs = model(input_ids)

        # Mean pooling of embeddings for each sentence in the batch
        batch_embeddings = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
        embeddings.extend(batch_embeddings)

    return embeddings

# Apply the function to get embeddings for the 'tags' column in batches
train_sample['bert_embedding'] = get_distilbert_embeddings_batch(train_sample['tags'].tolist(), batch_size=16, max_length=64)

# Display the results
train_sample[['tags', 'bert_embedding']].head()
```

**Using Self-Attention**


```python
from transformers import DistilBertModel, DistilBertTokenizer
import torch
import numpy as np
import pandas as pd

# Check if GPU is available and load model to GPU if possible
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the DistilBERT model and tokenizer
model_name = "distilbert-base-uncased"
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DistilBertModel.from_pretrained(model_name, output_attentions=True).to(device)

# Define a function to get DistilBERT embeddings with attention-weighted pooling in batches
def get_attention_weighted_embeddings_batch(texts, batch_size=16, max_length=64):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]

        # Tokenize the batch of texts
        inputs = tokenizer(batch_texts, return_tensors="pt", truncation=True, padding=True, max_length=max_length)
        input_ids = inputs['input_ids'].to(device)
        attention_mask = inputs['attention_mask'].to(device)

        # Generate embeddings and attention weights for the batch
        with torch.no_grad():
            outputs = model(input_ids, attention_mask=attention_mask)
            last_hidden_state = outputs.last_hidden_state  # Shape: (batch_size, seq_len, hidden_size)
            attention_weights = outputs.attentions[-1]  # Get the attention weights from the last layer

        # Average across attention heads
        attention_weights = attention_weights.mean(dim=1)  # Shape: (batch_size, seq_len, seq_len)

        # Expand attention_mask to match attention_weights' dimensions
        expanded_mask = attention_mask.unsqueeze(-1).float()  # Shape: (batch_size, seq_len, 1)
        expanded_attention_weights = attention_weights * expanded_mask  # Mask padded positions

        # Normalize weights to sum to 1 over valid tokens
        sum_attention_weights = expanded_attention_weights.sum(dim=1, keepdim=True) + 1e-9
        normalized_attention_weights = expanded_attention_weights / sum_attention_weights

        # Compute weighted sum of token embeddings for each sentence
        weighted_embeddings = torch.bmm(normalized_attention_weights.transpose(1, 2), last_hidden_state)
        sentence_embeddings = weighted_embeddings.squeeze(1)  # Shape: (batch_size, hidden_size)

        # Convert to numpy and add to the results
        batch_embeddings = sentence_embeddings.cpu().numpy()
        embeddings.extend(batch_embeddings)

    return embeddings

# Apply the function to get embeddings for the 'tags' column with attention-weighted pooling
train_sample['bert_embedding'] = get_attention_weighted_embeddings_batch(
    train_sample['tags'].tolist(), batch_size=16, max_length=64
)

# Display the results
train_sample[['tags', 'bert_embedding']].head()
```

#Clubbing it with cosine similarity for getting better output

```python
from transformers import DistilBertModel, DistilBertTokenizer
import torch
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
from transformers import DistilBertModel, DistilBertTokenizer
import torch
import numpy as np
import pandas as pd

# Check if GPU is available and load model to GPU if possible
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the DistilBERT model and tokenizer
model_name = "distilbert-base-uncased"
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DistilBertModel.from_pretrained(model_name, output_attentions=True).to(device)

# Function to get average pooled embeddings
def get_average_pooling_embeddings_batch(texts, batch_size=16, max_length=64):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]

        # Tokenize the batch of texts
        inputs = tokenizer(batch_texts, return_tensors="pt", truncation=True, padding=True, max_length=max_length)
        input_ids = inputs['input_ids'].to(device)
        attention_mask = inputs['attention_mask'].to(device)

        # Generate embeddings for the batch
        with torch.no_grad():
            outputs = model(input_ids, attention_mask=attention_mask)
            last_hidden_state = outputs.last_hidden_state  # Shape: (batch_size, seq_len, hidden_size)

        # Apply average pooling across the sequence length dimension
        sentence_embeddings = last_hidden_state.mean(dim=1)  # Shape: (batch_size, hidden_size)

        # Convert to numpy and add to the results
        batch_embeddings = sentence_embeddings.cpu().numpy()
        embeddings.extend(batch_embeddings)

    return embeddings

# Apply the function to get embeddings for the 'tags' column with average pooling
train_sample['bert_embedding'] = get_average_pooling_embeddings_batch(
    train_sample['tags'].tolist(), batch_size=16, max_length=64
)

# Ensure 'bert_embedding' in train_sample is in 2D (for cosine similarity)
train_sample['bert_embedding'] = train_sample['bert_embedding'].apply(lambda x: np.array(x).flatten())

# PCA: Reduce the dimensionality of the embeddings to 768 (common size for BERT-based models)
pca = PCA(n_components=768)

# Reduce stored embeddings to 768 dimensions
train_sample_embeddings = np.vstack(train_sample['bert_embedding'].to_numpy())
train_sample_reduced_embeddings = pca.fit_transform(train_sample_embeddings)

# Function to get embedding for a new input question
def get_question_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=64)
    input_ids = inputs['input_ids'].to(device)
    attention_mask = inputs['attention_mask'].to(device)

    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
        last_hidden_state = outputs.last_hidden_state

        # Apply average pooling
        sentence_embedding = last_hidden_state.mean(dim=1).squeeze(0).cpu().numpy()  # Reduced to 1D

    return sentence_embedding

```

```python
# Define the function to find similar questions
def find_similar_questions(input_text, top_n=5):
    # Get embedding for the input text
    input_embedding = get_question_embedding(input_text)

    # Reduce the dimensionality of the input embedding to match the stored embeddings
    input_embedding_reduced = pca.transform(input_embedding.reshape(1, -1))

    # Calculate cosine similarity with each question embedding in train_sample
    similarities = cosine_similarity(input_embedding_reduced, train_sample_reduced_embeddings)

    # Get the top N most similar questions
    top_indices = similarities[0].argsort()[-top_n:][::-1]
    similar_questions = train_sample.iloc[top_indices][['name']]

    return similar_questions
```

```python
# Example usage
input_question = "Merge Two Sorted Lists"
similar_questions = find_similar_questions(input_question, top_n=20)
similar_questions
```
