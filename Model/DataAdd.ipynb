```python
import numpy as np # linear algebra
import pandas as pd
import json

# Assuming the file is named 'data.json'
file_name = './preprocessed_data.json'

# Open and load the JSON file
with open(file_name, 'r') as file:
    data = json.load(file)
```

```python
import pandas as pd

# Normalize the JSON data into a flat table
df = pd.json_normalize(data)

# Display the DataFrame
df.columns
```

```python
import pandas as pd

# Convert the JSON data to a DataFrame
df = pd.DataFrame(data)
# Set 'id' as the index
df = df.set_index('id')
# Reset index and drop the default index column
df = df.reset_index(drop=True)

# Display the DataFrame
df.sample(1)
```

```python
import requests

# API endpoint
url = "http://localhost:3000/api/questions"

try:
    response = requests.delete(url)
    
    if response.status_code == 200:
        print("All questions deleted successfully!")
    else:
        print(f"Failed to delete questions. Status code: {response.status_code}")
        print("Response:", response.text)
except requests.exceptions.RequestException as e:
    print("An error occurred:", e)
```

```python
import pandas as pd
import requests
import random
import time  # Import time module for adding delay


# API endpoint
url = "http://localhost:3000/api/questions"

# Read the CSV file into a DataFrame
questions_df = df  # Assuming `df` is the DataFrame containing your question data

# Iterate over each row in the DataFrame
for index, row in questions_df.iterrows():
    # Prepare the payload in the required format
    payload = {
        "qId": index + 1,  # Assign unique ID starting from 1
        "tags": row['topics'].split(','),  # Split topics into a list
        "name": row['name'],
        "link": row['link'],
        "totalSolved": row['accecpted'],  # Assuming this maps to totalSolved
        "upvotes": random.randint(100, 5000),  # Generate random upvotes
        "downvotes": random.randint(10, 500),  # Generate random downvotes
        "difficulty": row['difficulty'],
        "avgTime": random.randint(20, 120)  # Generate random average time
    }

    try:
        # Send the POST request
        response = requests.post(url, json=payload)
        
        # Check the response status
        if response.status_code == 201:  # Assuming 201 is the success status code
            print(f"Question {index + 1} added successfully!")
        else:
            print(f"Failed to add Question {index + 1}. Status code: {response.status_code}")
            print("Response:", response.text)
        
        # time.sleep(1)  # Adjust the sleep time (in seconds) as necessary
    
    except requests.exceptions.RequestException as e:
        print(f"An error occurred while adding Question {index + 1}: {e}")
        # time.sleep(5)  
```
